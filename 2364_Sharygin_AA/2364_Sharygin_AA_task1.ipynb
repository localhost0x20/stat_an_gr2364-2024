## Основные теоретические положения

Нормальное распределение (одномерный случай):

$ \xi \sim \mathcal{N}(a, \sigma^2) $

$ p_\xi (x) = \cfrac{1}{\sigma\sqrt{2\pi}} \cdot \exp \left(-\cfrac{1}{2}\left(\cfrac{x-a}{\sigma}\right)^2\right) $

Многомерное нормальное распределение:

$ \overrightarrow{\xi} \sim \mathcal{N}(\overrightarrow{\mu}, \Sigma) $

$ p_{\overrightarrow{\xi}} \left(\overrightarrow{x}\right) = \cfrac{1}{\left(2\pi\right)^{n/2} \sqrt{\det\Sigma}} \cdot \exp \left(-\cfrac{1}{2}\left(\overrightarrow{x}-\mu\right)^{\intercal}\Sigma^{-1}\left(\overrightarrow{x}-\mu\right)\right) $

Преобразование характеристик:

$ \overrightarrow{\xi} \sim \mathcal{N}(\overrightarrow{\mu}, \Sigma) $

$ \overrightarrow{\eta} = A \overrightarrow{\xi} + \overrightarrow{b} $

$ \overrightarrow{\eta} \sim \mathcal{N}(A \overrightarrow{\mu} + \overrightarrow{b}, A \Sigma A^{\intercal}) $

Условное распределение для абсолютно непрерывных величин:

$ p_{\xi|\eta=y_0} (x) = \cfrac{p_{\xi, \eta}  (x, y_0)}{p_\eta  (y_0)} $

Стандартное нормально распределение - нормальное распределение с нулевым мат. ожиданием и единичной дисперсией. В многомерном случае - нулевой вектор мат. ожидания и единичная ковариационная матрица.

Центрированный случайный вектор - случайный вектор с нулевым мат. ожиданием.

Аффинное преобразование - преобразование вида:

$ \begin{pmatrix}
x\\
y
\end{pmatrix} = A \begin{pmatrix}
u\\
v
\end{pmatrix} + \begin{pmatrix}
\alpha\\
\beta
\end{pmatrix} $

**Плотность двумерного нормального распределения имеет вид:**

$ p_{\xi, \eta}  (x, y) = C \cdot \exp \left(-\cfrac{1}{2}\left(3x^2+3xy+7y^2-9x-17y+13\right)\right) $

**Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $ p_{\xi, \eta} $ .**

Многомерное нормальное распределение:

$ \overrightarrow{\xi} \sim \mathcal{N}(\overrightarrow{\mu}, \Sigma) $

$ p_{\overrightarrow{\xi}} \left(\overrightarrow{x}\right) = \cfrac{1}{\left(2\pi\right)^{n/2} \sqrt{\det\Sigma}} \cdot \exp \left(-\cfrac{1}{2}\left(\overrightarrow{x}-\mu\right)^{\intercal}\Sigma^{-1}\left(\overrightarrow{x}-\mu\right)\right) $

Преобразуем исходное выражение и вычислим вектор математических ожиданий $ \overrightarrow{\mu} $ и ковариационную матрицу $ \Sigma $:

$ p_{\xi, \eta}  (x, y) = C \cdot \exp \left(-\cfrac{1}{2}\left(3x^2+3xy+7y^2-9x-17y+13\right)\right) = C \cdot \exp \left(-\cfrac{1}{2}\left(3 \left(x^2+xy+\cfrac{1}{4}y^2\right)+\cfrac{25}{4}y^2-9x-17y+13\right)\right) = $

$ = C \cdot \exp \left(-\cfrac{1}{2}\left(3 \left(x+\cfrac{1}{2}y\right)^2+\cfrac{25}{4}y^2-9x-17y+13\right)\right) = C \cdot \exp \left(-\cfrac{1}{2}\left(3 \left(x+\cfrac{1}{2}y\right)^2+\cfrac{25}{4}y^2-9 \left(x+\cfrac{1}{2}y\right)-\cfrac{25}{2}y+13\right)\right) = $

$ = C \cdot \exp \left(-\cfrac{1}{2}\left(3 \left(\left(x+\cfrac{1}{2}y\right)^2-3 \left(x+\cfrac{1}{2}y\right)+\cfrac{9}{4}\right)+\cfrac{25}{4}y^2-\cfrac{25}{2}y+\cfrac{25}{4}\right)\right) = C \cdot \exp \left(-\cfrac{1}{2}\left(3 \left(x+\cfrac{1}{2}y-\cfrac{3}{2}\right)^2+\cfrac{25}{4} \left(y-1\right)^2\right)\right) = $

$ = C \cdot \exp \left(-\cfrac{1}{2}\left(3 \left(\left(x-1\right)+\cfrac{1}{2}\left(y-1\right)\right)^2+\cfrac{25}{4} \left(y-1\right)^2\right)\right) = C \cdot \exp \left(-\cfrac{1}{2}\left(3\left(x-1\right)^2+3\left(x-1\right)\left(y-1\right)+7\left(y-1\right)^2\right)\right) $

Представим в виде матричного умножения:

$ p_{\xi, \eta}  (x, y) = C \cdot \exp \left(-\cfrac{1}{2}\left(x-1, y-1\right)\begin{pmatrix}
3 & \ 3/2\\
3/2 & 7
\end{pmatrix}\begin{pmatrix}
x-1\\
y-1
\end{pmatrix}\right) $

Отсюда видно, что вектор математических ожиданий равен:

$ \mathbb{E} \begin{pmatrix}

$ \Sigma^{-1} = \begin{pmatrix}

$ \Sigma = \cfrac{4}{75}\begin{pmatrix}
7 & \ -3/2\\
-3/2 & 3
\end{pmatrix} $

Найдём дисперсии компонент:

$ \mathbb{D} \xi = \cfrac{28}{75} $

$ \mathbb{D} \eta = \cfrac{12}{75} = \cfrac{4}{25} $

Теперь найдём ковариацию и коэффициент корреляции:

$ \mathrm{cov} (\xi, \eta) = -\cfrac{2}{25} $

$ \rho (\xi, \eta) = \cfrac{\mathrm{cov} (\xi, \eta)}{\sqrt{\mathbb{D} \xi \cdot \mathbb{D} \eta}} = -\cfrac{2 \cdot 75}{4 \cdot 25\sqrt{21}} = -\cfrac{3}{2\sqrt{21}} \thickapprox -0.327 $

Найдём константу $ C $:

$ C = \cfrac{1}{\left(2\pi\right)^{n/2} \sqrt{\det\Sigma}} = \cfrac{5\sqrt{3}}{4\pi} $

Итоговая функция распределения:

$ p_{\xi, \eta}  (x, y) = \cfrac{5\sqrt{3}}{4\pi} \cdot \exp \left(-\cfrac{1}{2}\left(3x^2+3xy+7y^2-9x-17y+13\right)\right) $

Нормальное распределение (одномерный случай):

$ \xi \sim \mathcal{N}(a, \sigma^2) $

$ p_\xi (x) = \cfrac{1}{\sigma\sqrt{2\pi}} \cdot \exp \left(-\cfrac{1}{2}\left(\cfrac{x-a}{\sigma}\right)^2\right) $

Зная характеристики случайного распределения $ \eta $ (из первой задачи), найдём функцию распределения:

$ \mathbb{E} \eta = 1, \mathbb{D} \eta = \cfrac{4}{25} $

$ p_\eta (y) = \cfrac{5}{2\sqrt{2\pi}} \cdot \exp \left(-\cfrac{1}{2}\left(\cfrac{y-1}{2/5}\right)^2\right) = \cfrac{5}{2\sqrt{2\pi}} \cdot \exp \left(-\cfrac{1}{2}\left(\cfrac{25}{4}y^2-\cfrac{25}{2}y+\cfrac{25}{4}\right)\right) $

Подставим и преобразуем:

$ p_{\xi|\eta=y_0} (x) = \cfrac{\cfrac{5\sqrt{3}}{4\pi} \cdot \exp \left(-\cfrac{1}{2}\left(3x^2+3xy_0+7y_0^2-9x-17y_0+13\right)\right)}{\cfrac{5}{2\sqrt{2\pi}} \cdot \exp \left(-\cfrac{1}{2}\left(\cfrac{25}{4}y_0^2-\cfrac{25}{2}y_0+\cfrac{25}{4}\right)\right)} = \cfrac{2\sqrt{3}\sqrt{2\pi}}{4\pi} \cdot \exp \left(-\cfrac{1}{2}\left(3x^2+3xy_0+\cfrac{3}{4}y_0^2-9x-\cfrac{9}{2}y_0+\cfrac{27}{4}\right)\right) = $

$ = \cfrac{\sqrt{3}}{\sqrt{2\pi}} \cdot \exp \left(-\cfrac{1}{2}\left(3x^2+3\left(y_0-3\right)x+\cfrac{3}{4}\left(y_0^2-6y_0+9\right)\right)\right) = \cfrac{\sqrt{3}}{\sqrt{2\pi}} \cdot \exp \left(-\cfrac{1}{2} \cdot 3\left(x^2+\left(y_0-3\right)x+\cfrac{1}{4}\left(y_0-3\right)^2\right)\right) = $

$ = \cfrac{\sqrt{3}}{\sqrt{2\pi}} \cdot \exp \left(-\cfrac{1}{2} \cdot 3\left(x+\cfrac{y_0-3}{2}\right)^2\right) = \cfrac{\sqrt{3}}{\sqrt{2\pi}} \cdot \exp \left(-\cfrac{1}{2} \left(\cfrac{x+\cfrac{y_0-3}{2}}{1/\sqrt{3}}\right)^2\right) $

$ p_{\xi|\eta=y_0} (x) = \cfrac{\sqrt{3}}{\sqrt{2\pi}} \cdot \exp \left(-\cfrac{1}{2} \cdot 3\left(x+\cfrac{y_0-3}{2}\right)^2\right) $

Отсюда видно:

$ \mathbb{E} \left(\xi|\eta\right) = -\cfrac{y_0-3}{2} $

$ \mathbb{D} \left(\xi|\eta\right) = \cfrac{1}{3} $
